```python
import collections
import pandas as pd
import functools
from functools import lru_cache
import sympy
```

```python
with open('../inputs/input21.txt', 'r') as f:
    lines = f.readlines()
n = len(lines)
for i in range(n):
    lines[i] = lines[i].strip()
    lines[i] = lines[i].split()
    lines[i][0] = lines[i][0][:4]
    if lines[i][1].isnumeric():
        lines[i][1] = int(lines[i][1])
```

```python
monkey_map = {}
for line in lines:
    monkey_map[line[0]] = line[1:]
```

```python
def dfs(node):
    val = monkey_map[node]
    if isinstance(val[0], int):
        return val[0]
    elif val[1]=='+':
        return dfs(val[0]) + dfs(val[2])
    elif val[1]=='-':
        return dfs(val[0]) - dfs(val[2])
    elif val[1]=='*':
        return dfs(val[0]) * dfs(val[2])
    elif val[1]=='/':
        return dfs(val[0]) // dfs(val[2])
```

```python
#Answer to Part 1
dfs('root')
```

```python
def contains_humn(node):
    if node=='humn':
        return True
    val = monkey_map[node]
    if isinstance(val[0], int):
        return False
    else:
        return contains_humn(val[0]) or contains_humn(val[2])
```

```python
print(contains_humn('jhpn'))
```

```python
#The number to match
dfs('jmsg')
```

```python
def dfs2(node):
    if node=='humn':
        return "x"
    val = monkey_map[node]
    if isinstance(val[0], int):
        return str(val[0])
    else:
        return "(" + dfs2(val[0]) + val[1] + dfs2(val[2]) + ")"
```

```python
equation = str(dfs('jmsg'))+"="+dfs2('jhpn')
print(equation)
```

```python
sympy_eq = sympy.sympify("Eq(" + equation.replace("=", ",") + ")")
sympy.solve(sympy_eq)[0]
```

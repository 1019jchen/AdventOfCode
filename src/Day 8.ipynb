```python
with open('input.txt', 'r') as f:
    lines = f.readlines()
n = len(lines)
for i in range(n):
    lines[i] = lines[i].strip()
    lines[i] = list(lines[i])
for i in range(n):
    for j in range(n):
        lines[i][j] = int(lines[i][j])
print(lines)
```

```python
n = len(lines)
m = len(lines[0])
print(n,m)
```

```python
visible = [[False]*99 for i in range(99)]
for i in range(n):
    for j in range(n):
        tree = lines[i][j]
        #check above:
        vis = True
        for x in range(0,i):
            if tree <= lines[x][j]:
                vis = False
        if vis == True:
            visible[i][j] = True
        #check left:
        vis = True
        for x in range(0,j):
            if tree <= lines[i][x]:
                vis = False
        if vis == True:
            visible[i][j] = True
        #check down:
        vis = True
        for x in range(i+1,n):
            if tree <= lines[x][j]:
                vis = False
        if vis == True:
            visible[i][j] = True
        #check right:
        vis = True
        for x in range(j+1,n):
            if tree <= lines[i][x]:
                vis = False
        if vis == True:
            visible[i][j] = True
print(visible)
```

```python
count = 0
for i in range(n):
    for j in range(n):
        if visible[i][j]:
            count += 1
print(count)
```

```python
score = [[0]*n for i in range(n)]
for i in range(n):
    for j in range(n):
        tree = lines[i][j]
        #check above:
        above = 0
        for x in range(i-1,-1,-1):
            if tree > lines[x][j]:
                above += 1
            else:
                above += 1
                break
        #check left:
        left = 0
        for x in range(j-1,-1,-1):
            if tree > lines[i][x]:
                left += 1
            else:
                left += 1
                break
        #check down:
        down = 0
        for x in range(i+1,n):
            if tree > lines[x][j]:
                down += 1
            else:
                down += 1
                break
        #check right:
        right = 0
        for x in range(j+1,n):
            if tree > lines[i][x]:
                right += 1
            else:
                right += 1
                break
        if i==3 and j==2:
            print(above,left,down,right)
        score[i][j] = above * left * down * right
print(score)
```

```python
maxScore = 0
for i in range(n):
    for j in range(n):
        if score[i][j]>maxScore:
            maxScore = score[i][j]
print(maxScore)
```

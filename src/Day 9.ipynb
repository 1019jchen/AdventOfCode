```python
with open('../inputs/input9.txt', 'r') as f:
    lines = f.readlines()
n = len(lines)
for i in range(n):
    lines[i] = lines[i].split()
    lines[i][1] = int(lines[i][1])
print(lines)
```

```python
#Initial coordinates:
headX = 0
headY = 0
tailX = 0
tailY = 0
visited = set()
```

```python
directions = {'U':[0,1], 'R': [1,0], 'D': [0,-1], 'L': [-1,0]}
for line in lines:
    for i in range(line[1]):
        headX += directions[line[0]][0]
        headY += directions[line[0]][1]
        distX = headX - tailX
        distY = headY - tailY
        #tail needs to move right
        if distX>=2 and distY == 0:
            tailX += 1
        #tail needs to move left
        elif distX<=-2 and distY ==0:
            tailX -= 1
        #tail needs to move up
        elif distX==0 and distY>=2:
            tailY += 1
        #tail needs to move down
        elif distX == 0 and distY<=-2:
            tailY -= 1
        #tail needs to move up and to the right
        elif (distX>=2 and distY>=1) or (distX>=1 and distY>=2):
            tailX += 1
            tailY += 1
        #tail needs to move up and to the left
        elif (distX<=-1 and distY>=2) or (distX<=-2 and distY>=1):
            tailX -= 1
            tailY += 1
        #tail needs to move down and to the left
        elif (distX<=-1 and distY<=-2) or (distX<=-2 and distY<=-1):
            tailX -= 1
            tailY -= 1
        #tail needs to move down and to the right
        elif (distX>=2 and distY<=-1) or (distX>=1 and distY<=-2):
            tailX += 1
            tailY -= 1
        visited.add((tailX, tailY))
visited
```

```python

#List to represent the coordinates of each knot
knots = [(0,0) for i in range(10)]
knots
```

```python
#Reusing previous part as a function
def shift(headX, headY, tailX, tailY):
    distX = headX - tailX
    distY = headY - tailY
    #tail needs to move right
    if distX>=2 and distY == 0:
        tailX += 1
    #tail needs to move left
    elif distX<=-2 and distY ==0:
        tailX -= 1
    #tail needs to move up
    elif distX==0 and distY>=2:
        tailY += 1
    #tail needs to move down
    elif distX == 0 and distY<=-2:
        tailY -= 1
    #tail needs to move up and to the right
    elif (distX>=2 and distY>=1) or (distX>=1 and distY>=2):
        tailX += 1
        tailY += 1
    #tail needs to move up and to the left
    elif (distX<=-1 and distY>=2) or (distX<=-2 and distY>=1):
        tailX -= 1
        tailY += 1
    #tail needs to move down and to the left
    elif (distX<=-1 and distY<=-2) or (distX<=-2 and distY<=-1):
        tailX -= 1
        tailY -= 1
    #tail needs to move down and to the right
    elif (distX>=2 and distY<=-1) or (distX>=1 and distY<=-2):
        tailX += 1
        tailY -= 1
    return (tailX,tailY)
```

```python
visited = set()
for line in lines:
    for i in range(line[1]):
        knots[0] = (knots[0][0] + directions[line[0]][0], knots[0][1] + directions[line[0]][1])
        for j in range(9):
            knots[j+1] = shift(knots[j][0], knots[j][1], knots[j+1][0], knots[j+1][1])
        visited.add(knots[-1])
```

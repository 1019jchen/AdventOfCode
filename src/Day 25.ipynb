```python
import collections
import pandas as pd
import functools
from functools import lru_cache
import sympy
import numpy as np
```

```python
with open('../inputs/input25.txt', 'r') as f:
    lines = f.readlines()
n = len(lines)
for i in range(n):
    lines[i] = lines[i].strip()
lines
```

```python
conversions = {'1':1, '2':2, '0':0, "-":-1, "=":-2}
normal_nums = []
```

```python
def snafu_to_int(snafu):
    num = 0
    place = 1
    pointer = len(snafu)-1
    while (pointer>=0):
        num += conversions[snafu[pointer]]*place
        place *= 5
        pointer -= 1
    return num
```

```python
for line in lines:
    normal_nums.append(snafu_to_int(line))
normal_nums
```

```python
total_fuel = sum(normal_nums)
total_fuel
```

```python
#Answer to Part 1
snafu = ""
while (total_fuel > 0):
    place = total_fuel%5
    if place<3:
        snafu = str(place)+snafu
        total_fuel = total_fuel // 5
    elif place == 3:
        snafu = "=" + snafu
        total_fuel = (total_fuel // 5) + 1
    elif place == 4:
        snafu = "-" + snafu
        total_fuel = (total_fuel // 5)+1
print(snafu)
```

```python
import collections
import pandas as pd
import ast
import functools
```

```python
class Node():
    def __init__(self):
        self.parent = None
        self.children = []
```

```python
with open('../inputs/input13.txt', 'r') as f:
    lines = f.readlines()
n = len(lines)
nodes = []
for i in range(n):
    if (i+1)%3==0:
        continue
    root = Node()
    cur = root
    for j in range(1,len(lines[i])-1):
        if lines[i][j] == '[':
            temp = Node()
            temp.parent=cur
            cur.children.append(temp)
            cur = temp
        elif lines[i][j] == ']':
            cur = cur.parent
        elif lines[i][j] == '1':
            if lines[i][j+1]=='0':
                cur.children.append(10)
                continue
        elif lines[i][j]=='0':
            if lines[i][j-1]=='1':
                continue
        if lines[i][j] in '1234567890':
            cur.children.append(int(lines[i][j]))
    nodes.append(root)
```

```python
def compareNodes(left, right):
    l = left.children
    r = right.children
    for element1, element2 in zip(l,r):
        if isinstance(element1, int) and isinstance(element2, int):
            if element1<element2:
                return -1
            elif element1>element2:
                return 1
            continue
        elif isinstance(element1, int) or isinstance(element2, int):
            if isinstance(element1, int):
                temp = Node()
                temp.children.append(element1)
                element1 = temp
            if isinstance(element2, int):
                temp = Node()
                temp.children.append(element2)
                element2 = temp
        result = compareNodes(element1, element2)
        if result:
            return result
    if len(l)<len(r):
        return -1
    elif len(l)>len(r):
        return 1
    else:
        return 0
    
```

```python
m = len(nodes)
orders = [0]
for i in range(0,m,2):
    orders.append(compareNodes(nodes[i],nodes[i+1]))
orders
```

```python
#Answer to Part 1
answer = 0
for i in range(len(orders)):
    if orders[i]==-1:
        answer+=i
answer
```

```python
#Divider packet 1
packet1 = Node()
temp = Node()
temp.children.append(2)
packet1.children.append(temp)

#Divider packet 2
packet2 = Node()
temp = Node()
temp.children.append(6)
packet2.children.append(temp)

nodes.append(packet1)
nodes.append(packet2)
```

```python
sorted_nodes = sorted(nodes, key=functools.cmp_to_key(compareNodes))
```

```python
for i in range(len(sorted_nodes)):
    node = sorted_nodes[i]
    if len(node.children)==1 and not isinstance(node.children[0], int):
        if len(node.children[0].children)==1 and node.children[0].children[0]==2:
            index1 = i+1
    if len(node.children)==1 and not isinstance(node.children[0], int):
        if len(node.children[0].children)==1 and node.children[0].children[0]==6:
            index2 = i+1
```

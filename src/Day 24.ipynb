```python
import collections
import pandas as pd
import functools
from functools import lru_cache
import sympy
import numpy as np
```

```python
with open('../inputs/input24.txt', 'r') as f:
    lines = f.readlines()
n = len(lines)
for i in range(n):
    lines[i] = list(lines[i].strip())
    m = len(lines[0])
lines
```

```python
blizzards = [[0]*(m-2) for x in range(n-2)]
for i in range(n-2):
    for j in range(m-2):
        blizzards[i][j] = lines[i+1][j+1]
blizzards
```

```python
up_blizzards = set()
down_blizzards = set()
right_blizzards = set()
left_blizzards = set()
n = n-2
m = m-2
for i in range(n):
    for j in range(m):
        if blizzards[i][j] == '>':
            right_blizzards.add((i,j))
        elif blizzards[i][j] == '<':
            left_blizzards.add((i,j))
        elif blizzards[i][j] == '^':
            up_blizzards.add((i,j))
        elif blizzards[i][j] == 'v':
            down_blizzards.add((i,j))
```

```python
def check_right_blizz(time):
    res = set()
    for b in right_blizzards:
        res.add((b[0],(b[1]+time)%m))
    return res
```

```python
def check_left_blizz(time):
    res = set()
    for b in left_blizzards:
        res.add((b[0],(b[1]-time)%m))
    return res
```

```python
def check_up_blizz(time):
    res = set()
    for b in up_blizzards:
        res.add(((b[0]-time)%n, b[1]))
    return res
```

```python
def check_down_blizz(time):
    res = set()
    for b in down_blizzards:
        res.add(((b[0]+time)%n, b[1]))
    return res
```

```python
#Answer to Part 1
x,y = -1,0
time = 0
Q = set()
#queue contains coordinates
Q.add((x,y))
not_reached = True
directions = [(-1, 0), (1, 0), (0, -1), (0, 1), (0, 0)]
start = (-1, 0)
destination = (n,m-1)
route_found = False

while (not route_found):
    next_Q = set()
    time += 1
    temp_blizz = check_left_blizz(time).union(check_right_blizz(time), check_down_blizz(time), check_up_blizz(time))
    for coords in Q:
        for d in directions:
            temp_coords = (coords[0] + d[0], coords[1] + d[1])
            if temp_coords not in temp_blizz:
                if temp_coords == destination:
                    not_reached = False
                    print(time)
                    route_found = True
                elif temp_coords == start or (temp_coords[0] >= 0 and temp_coords[0] < n and temp_coords[1] >= 0 and temp_coords[1] < m):
                    next_Q.add(temp_coords)
    Q = next_Q.copy()
```
